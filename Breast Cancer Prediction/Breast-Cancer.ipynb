```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bc = pd.read_csv('data.csv')
bc.head()
```

```python
bc.drop(['id'],axis=1,inplace=True)
```

```python
bc.isnull().sum()
```

```python
bc.dtypes
```

```python
bc.describe()
```

```python
bc.corr(numeric_only = True)
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x=bc['diagnosis'].value_counts().index,y=bc['diagnosis'].value_counts().values)
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(bc.corr(numeric_only = True),annot=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(bc.drop(['diagnosis'],axis=1),bc['diagnosis'],test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

```python
y_pred = dtree.predict(X_test)
```

```python
print('Predicted values: ',y_pred[:10])
print('Actual values: ',y_test[:10])
```

```python
print(dtree.score(X_test,y_test))
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
yhat = logmodel.predict(X_test)
```

```python
print('Predicted values: ',yhat[:10])
print('Actual values: ',y_test[:10])
```
